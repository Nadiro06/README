```
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer

# Cargar dataset
df = pd.read_csv('06 - dataset.csv')
```

### Imputación de valores faltantes

```
# Imputar variables numéricas con la media
num_cols = df.select_dtypes(include='number').columns
imputer_num = SimpleImputer(strategy='mean')
df[num_cols] = imputer_num.fit_transform(df[num_cols])
```

```
# Imputar variables categóricas con la moda
cat_cols = df.select_dtypes(include='object').columns
imputer_cat = SimpleImputer(strategy='most_frequent')
df[cat_cols] = imputer_cat.fit_transform(df[cat_cols])
```

### Codificación de variables categóricas

```
# Codificar todas las variables categóricas con LabelEncoder
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

### Guardar dataset preprocesado

```
df.to_csv('06 - dataset.csv', index=False)
```
